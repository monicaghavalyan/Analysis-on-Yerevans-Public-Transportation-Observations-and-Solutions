```python
import os
import osmnx as ox
```

```python
if os.path.exists('/content/drive/My Drive/Capstone/Data'):
    base_dir = '/content/drive/My Drive/Capstone/Data'
else:
    base_dir = os.path.abspath(os.path.join(os.getcwd(), '..', 'Data'))
```

```python
ox.settings.log_console = True
ox.settings.use_cache = True

place = "Yerevan, Armenia"

gdf_boundary = ox.geocode_to_gdf(place)

tags = {'building': True}
gdf_buildings = ox.features_from_place(place, tags)
columns_to_keep = ['building', 'geometry']
gdf_clean = gdf_buildings[columns_to_keep].copy()

G_walk = ox.graph_from_place(place, network_type='walk')
nodes, edges = ox.graph_to_gdfs(G_walk)
```

```python
gdf_boundary.to_file(os.path.join(base_dir, "source/yerevan_boundary.gpkg"), layer='boundary', driver="GPKG")
```

```python
gdf_clean.to_file(os.path.join(base_dir, "source/yerevan_buildings.gpkg"), layer='buildings', driver="GPKG")
```

```python
nodes.to_file(os.path.join(base_dir, "source/yerevan_walk.gpkg"), layer='nodes', driver='GPKG')
edges.to_file(os.path.join(base_dir, "source/yerevan_walk.gpkg"), layer='edges', driver='GPKG')
```
